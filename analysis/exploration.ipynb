```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/tggates_clean.csv")

df.head()
```

```python
df.info()
```

```python
#Checking Toxic vs Non-Toxic Compounds
df["toxicity_label"].value_counts()

#plot
sns.countplot(data=df, x="toxicity_label")

plt.title("Toxic vs Non-Toxic Compounds")
plt.show()
```

```python
#Exploring Biomarker Genes
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(data=df, x="toxicity_label", y="alt", ax=axes[0])
axes[0].set_title("ALT vs Toxicity")

sns.boxplot(data=df, x="toxicity_label", y="ast", ax=axes[1])
axes[1].set_title("AST vs Toxicity")

plt.show()
```

```python
sns.boxplot(data=df, x="toxicity_label", y="cyp2e1")

plt.title("CYP2E1 Expression vs Toxicity")
```

#Interpretion: CYP2E1 is involved in drug metabolism and oxidative stress, often associated with liver injury.
